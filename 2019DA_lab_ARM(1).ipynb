### 예제 데이터셋

```python
dataset=[['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

### TransactionEncoder 함수: 트랜잭션 데이터를 one-hot array로 바꿔줌

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)
```

### Boolean 타입으로 encode된 one-hot array

```python
te_ary
```

### Int타입으로 encode된 one-hot array

```python
te_ary.astype("int")
```

### one-hot array의 칼럼은 트랜잭션의 아이템들

```python
te.columns_
```

### 아이템 목록

```python
te.columns_mapping_
```

```python
df=pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

### Apriori 알고리즘 실행하기

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.2)
```

### 아이템 셋(아이템 이름 명시)

```python
apriori(df, min_support=0.6, use_colnames=True)
```

### 빈발집합의 길이 

```python
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemsets
```

### 길이가 2이며, Support 값이 0.8인 빈발집합 찾기

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.8) ]
```

### 특정 빈발집합 조회
* 집합내 아이템 순서는 상관없음
* frozenset

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'Onion', 'Eggs'} ]
```

### 특정 아이템을 포함한 빈발집합 조회

```python
array=['Onion','Eggs']
frequent_itemsets.loc[frequent_itemsets['itemsets'].isin(array)]
```

### 연관규칙 생성 from 빈발집합

```python
from mlxtend.frequent_patterns import association_rules
```

### confindence 기준으로 cut-off

```python
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules
```

### 조건에 맞는 규칙 조회

```python
rules[ (rules['support'] >= 0.8) &
                   (rules['lift'] >= 0.8) ]
```

### 칼럼추가(antecedents 길이)

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
```
